```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

'''Studying limit cycle behavior of the Van Der Pol oscillator'''

mu = 4

def RHS(t,y,mu):
    X,Y = y

    dX = Y
    dY = mu*(1-X**2)*Y-X

    return np.array([dX,dY])

t = [0,200]

y0 = [10,10]
```

```python
sol = solve_ivp(lambda t,y:RHS(t,y,mu),t,y0,t_eval=np.linspace(0,200,20000))

plt.xlabel("Time Derivative of X")
plt.ylabel("Time Derivative of Y")
plt.plot(sol.y[0,:],sol.y[1,:])
```
