```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

t = [0,2000]
y0 = [30,15]
par = {"r1":1,"k1":7,"omega":6,"d":7,"j":0.5,"r2":0.1,"mu1":0,"mu2":0}

def HT(t,y,par):
    X,Y = y
    
    dX = par['r1'] * X * (1-X/par['k1']) - ((par['omega'] * X)/(par['d'] + par['d'] * X)) * Y - par['mu1'] * X

    dY = par['r2'] * Y * (1-(par['j'] * Y)/X) - par['mu2'] * Y

    
    return np.array([dX,dY])


```

```python
sol = solve_ivp(fun=lambda t,y:HT(t,y,par),t_span=t,y0=y0,t_eval=np.linspace(0,200,20000))

plt.plot(np.linspace(0,2000,20000),sol.y[0,:])
plt.plot(np.linspace(0,2000,20000),sol.y[1,:])
```
