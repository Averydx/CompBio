```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

y0 = [10] #[fish bears]
t = [0,100]

r = 0.5
k = 100

def LV(t,y,r,k):
    dx = r*y[0] * (1-y[0]/k)
    return np.array([dx])

```

```python
sol = solve_ivp(fun=lambda t,y :LV(t,y,r,k),t_span=t,y0=y0,t_eval=np.linspace(0,100,100))

plt.plot(sol.y[0,:])
```
