```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

y0 = [10,1] #[fish bears]
t = [0,50]


par = {"alpha":1.1,"beta":0.4,"delta":0.1,"gamma":0.4}
#y0 = [par['gamma']/par['delta'],par['alpha']/par['beta']]

def LV(t,y,par):
    dx = par['alpha'] * y[0] - par['beta'] * y[0] * y[1]
    dy = par['delta'] * y[0] * y[1] - par['gamma'] * y[1]

    return np.array([dx,dy])

```

```python
sol = solve_ivp(fun=lambda t,y :LV(t,y,par),t_span=t,y0=y0,t_eval=np.linspace(0,50,50000))

plt.plot(sol.y[0,:],label="Fish Population")
plt.plot(sol.y[1,:],label = "Bear Population")
plt.legend()
```
