```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

t = [0,5000]
y0 = [0.001,0.01]
par = {'r':0.01,'K':1,'A':1,'m':1,'s':0.01,'h':1}

def HT(t,y,par):
    X,Y = y
    
    dX = par['r'] * X*(1-X/par['K']) - (par['m'] * X * Y)/(par['A'] + X)
    dY = par['s'] * Y * (1-Y/(par['h']*X))
    
    return np.array([dX,dY])


```

```python
sol = solve_ivp(fun=lambda t,y:HT(t,y,par),t_span=t,y0=y0,t_eval=np.linspace(t[0],t[1],t[1]))

for i in range(2):
    plt.plot(sol.y[i,:])
```
