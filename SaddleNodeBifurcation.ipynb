```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

par = {"alpha":0.3}
t = [0,200]
rng = np.random.default_rng()
def Saddle(t,y,par):
    X,Y = y
    
    dX = par['alpha'] - X ** 2
    dY = -Y

    return np.array([dX,dY])


for i in range(100):
    y0 = [rng.uniform(0,1.5),rng.uniform(-1.5,1.5)]

    sol = solve_ivp(lambda t,y:Saddle(t,y,par),y0=y0,t_span=t,t_eval=np.linspace(0,200,20000))

    plt.plot(sol.y[0,:],sol.y[1,:],linewidth=0.5)
```
